## Filter_Bikes

This notebook follows Clean_Bikes. The clean images that we're interested in from the bikes dataset have a full white background. This notebook creates a csv containing only records for those images.

```python
import csv
import os.path

import skimage.transform
from skimage.io import imread
from PIL import Image
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

We expect the images to be in ../datasets/bikes_im/, and the pre-filter csv to be ../datasets/bikes.csv

```python
price_sheet = open('../datasets/bikes.csv', 'r')
```

```python
output_sheet = open('../datasets/bikes_filtered.csv', 'w', newline='')
writer = csv.writer(output_sheet)
```

```python
valid_rows = 0
i = -1

for row in csv.reader(price_sheet):
    i += 1
    if i % 100 == 0:
        print('row number: ' + str(i) + ', valid rows: ' + str(valid_rows))
    
    index = row[0]
    name = row[1]
    price = row[2]
    
    # load image corresponding to row
    try:
        im_path = '../datasets/bikes_im/' + str(index) + '.jpg'
        im = imread(im_path)

        # check percent of full white pixels in image
        pixels = im.reshape((-1, 3))
        bools = (pixels == 255)
        percent_white = np.sum(np.sum(bools, axis=1) == 3) / len(bools)
        if percent_white > 0.2:
            # valid image, write a row for it in output
            writer.writerow([index, name, price])
            valid_rows += 1
        
    except:
        print('Error when reading image.')
        continue
    
```

```python
price_sheet.close()
output_sheet.close()
```
